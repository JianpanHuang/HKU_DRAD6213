```python
#@title 1. Install and Import Required Libraries
# =========================================================
!pip install torch torchvision matplotlib numpy

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if device.type == 'cuda':
    print(f"GPU: {torch.cuda.get_device_name(0)}")

# Set Random Seeds for Reproducibility
def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)
```

```python
#@title 2. Load and Prepare MNIST Dataset
# =========================================================
# Define transforms
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # MNIST mean and std
])

# Download and load training data
train_dataset = torchvision.datasets.MNIST(
    root='./data',
    train=True,
    transform=transform,
    download=True
)

# Download and load test data
test_dataset = torchvision.datasets.MNIST(
    root='./data',
    train=False,
    transform=transform,
    download=True
)

# Split training data into train and validation (90% train, 10% validation)
train_size = int(0.9 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

print(f"\nTraining samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Test samples: {len(test_dataset)}")
```

```python
#@title 3. Create Data Loaders
# =========================================================
batch_size = 64

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=2
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2
)
```

```python
#@title 4. Visualize Sample Images
# =========================================================
def show_sample_images(loader, num_images=10):
    # Get a batch of images
    data_iter = iter(loader)
    images, labels = next(data_iter)

    # Create a grid of images
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    axes = axes.ravel()

    for i in range(num_images):
        axes[i].imshow(images[i].squeeze(), cmap='gray')
        axes[i].set_title(f'Label: {labels[i].item()}')
        axes[i].axis('off')

    plt.suptitle('Sample MNIST Images', fontsize=16)
    plt.tight_layout()
    plt.show()

show_sample_images(train_loader)
```

```python
#@title 5. Define the Neural Network Architecture
# =========================================================
import torch.nn as nn
class DigitRecognizerNN(nn.Module):
    def __init__(self):
        super(DigitRecognizerNN, self).__init__()
        # C1: Convlayer (1 in channel, 6 out channels, 5x5 kernel)
        self.conv1 = nn.Conv2d(1, 6, kernel_size=5)
        # p1: Average pooling (2x2)
        self.avgpool1 = nn.AvgPool2d(kernel_size=2, stride=2)
        # C2: Conv layer (6 in channels, 16 out channels, 5x5 kernel)
        self.conv2 = nn.Conv2d(6, 16, kernel_size=5)
        # p2: Average pooling (2x2)
        self.avgpool2 = nn.AvgPool2d(kernel_size=2, stride=2)
        # Fully connected layers
        # For MNIST input (28x28), after conv and pooling:
        # conv1: (28 - 5 + 1) = 24 -> pool -> 12
        # conv2: (12 - 5 + 1) = 8 -> pool -> 4
        # Final feature map size: 16 channels, 4x4
        self.fc1 = nn.Linear(16 * 4 * 4, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        # Input shape: (batch_size, 1, 28, 28)
        x = nn.functional.relu(self.conv1(x))
        x = self.avgpool1(x)
        x = nn.functional.relu(self.conv2(x))
        x = self.avgpool2(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = nn.functional.relu(self.fc1(x))
        x = nn.functional.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Create model instance
model = DigitRecognizerNN().to(device)
print(model)
print(f"\nTotal parameters: {sum(p.numel() for p in model.parameters())}")
```

```python
#@title 6. Define Loss Function and Optimizer
# =========================================================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)
```

```python
#@title 7. Training Function
# =========================================================
def train_epoch(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    progress_bar = tqdm(train_loader, desc='Training')
    for images, labels in progress_bar:
        images, labels = images.to(device), labels.to(device)

        # Initialize the gradients
        optimizer.zero_grad()

        # Forward propagation
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Back propagation and optimize
        loss.backward()
        optimizer.step()

        # Statistics
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        # Update progress bar
        progress_bar.set_postfix({
            'Loss': f'{running_loss/(len(progress_bar)):.4f}',
            'Acc': f'{100.*correct/total:.2f}%'
        })

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100. * correct / total
    return epoch_loss, epoch_acc
```

```python
#@title 8. Validation Function
# =========================================================
def validate(model, val_loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc='Validation'):
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    epoch_loss = running_loss / len(val_loader)
    epoch_acc = 100. * correct / total
    return epoch_loss, epoch_acc
```

```python
#@title 9. Training Loop
# =========================================================
num_epochs = 10
best_val_acc = 0
train_losses, val_losses = [], []
train_accs, val_accs = [], []

# Record the start time
import time
start_time = time.time()
print("Starting Training...")
print("=" * 50)

for epoch in range(num_epochs):
    print(f"\nEpoch [{epoch+1}/{num_epochs}]")

    # Training
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # Validation
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    # Learning rate scheduling
    scheduler.step(val_loss)

    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_mnist_model.pth')
        print(f"✓ New best model saved with validation accuracy: {val_acc:.2f}%")

    # Print epoch summary
    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
    print(f"Learning Rate: {optimizer.param_groups[0]['lr']:.6f}")

# Record the end time
end_time = time.time()
training_time = end_time - start_time
print("\n" + "=" * 50)
print(f"Training completed!")
print(f"Best validation accuracy: {best_val_acc:.2f}%")
print(f"Total training time: {training_time:.2f} seconds")
```

```python
#@title 10. Plot Training History
# ==================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot losses
ax1.plot(train_losses, label='Train Loss', marker='o')
ax1.plot(val_losses, label='Validation Loss', marker='o')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.set_title('Training and Validation Loss')
ax1.legend()
ax1.grid(True)

# Plot accuracies
ax2.plot(train_accs, label='Train Accuracy', marker='o')
ax2.plot(val_accs, label='Validation Accuracy', marker='o')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.set_title('Training and Validation Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
#@title 11. Test the Model
# ==================================================
# Load best model
model.load_state_dict(torch.load('best_mnist_model.pth'))

# Test the model
test_loss, test_acc = validate(model, test_loader, criterion, device)
print(f"\nTest Results:")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.2f}%")
```

```python
#@title 12. Plot Confusion Matrix
# ==================================================
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

def plot_confusion_matrix(model, test_loader, device):
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc='Computing predictions'):
            images = images.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)

            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())

    # Create confusion matrix
    cm = confusion_matrix(all_labels, all_predictions)

    # Plot
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(10), yticklabels=range(10))
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(all_labels, all_predictions, digits=4))

plot_confusion_matrix(model, test_loader, device)
```

```python
#@title 13. Visualize Predictions
# ==================================================
def visualize_predictions(model, test_loader, device, num_images=10):
    model.eval()

    # Get a batch of test images
    data_iter = iter(test_loader)
    images, labels = next(data_iter)
    images, labels = images[:num_images].to(device), labels[:num_images]

    # Get predictions
    with torch.no_grad():
        outputs = model(images)
        _, predicted = outputs.max(1)

    # Move images back to CPU for plotting
    images = images.cpu()

    # Plot images with predictions
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.ravel()

    for i in range(num_images):
        axes[i].imshow(images[i].squeeze(), cmap='gray')
        color = 'green' if predicted[i] == labels[i] else 'red'
        axes[i].set_title(f'True: {labels[i]}\nPred: {predicted[i]}', color=color)
        axes[i].axis('off')

    plt.suptitle(f'Model Predictions on Test Images (Green=Correct, Red=Incorrect)', fontsize=14)
    plt.tight_layout()
    plt.show()

    # Calculate accuracy for this batch
    batch_acc = (predicted.cpu() == labels).sum().item() / num_images * 100
    print(f"Batch Accuracy: {batch_acc:.2f}%")

visualize_predictions(model, test_loader, device)
```

```python
# 14. Visualize Misclassified Examples
# ==================================================
def show_misclassified(model, test_loader, device, num_examples=10):
    model.eval()
    misclassified_images = []
    misclassified_preds = []
    misclassified_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)

            # Find misclassified examples
            mask = predicted != labels
            misclassified_images.extend(images[mask].cpu())
            misclassified_preds.extend(predicted[mask].cpu())
            misclassified_labels.extend(labels[mask].cpu())

            if len(misclassified_images) >= num_examples:
                break

    # Plot misclassified examples
    num_examples = min(num_examples, len(misclassified_images))
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.ravel()

    for i in range(num_examples):
        axes[i].imshow(misclassified_images[i].squeeze(), cmap='gray')
        axes[i].set_title(f'True: {misclassified_labels[i]}\nPred: {misclassified_preds[i]}', color='red')
        axes[i].axis('off')

    plt.suptitle('Misclassified Examples', fontsize=14)
    plt.tight_layout()
    plt.show()

    print(f"Showing {num_examples} misclassified examples")

show_misclassified(model, test_loader, device)
```

```python
#@title 15. Model Inference Function
# ==================================================
def predict_digit(model, image_tensor, device):
    """
    Predict a single digit from a tensor image and return probabilities
    """
    model.eval()
    with torch.no_grad():
        image_tensor = image_tensor.unsqueeze(0).to(device)  # Add batch dimension
        output = model(image_tensor)
        probabilities = torch.softmax(output, dim=1)
        predicted_class = torch.argmax(probabilities, dim=1).item()
        confidence = probabilities[0][predicted_class].item()

    return predicted_class, confidence, probabilities[0].cpu().numpy()

# Test the inference function with a sample
sample_image, sample_label = test_dataset[20]
prediction, confidence, probabilities = predict_digit(model, sample_image, device)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot the digit image
ax1.imshow(sample_image.squeeze(), cmap='gray')
ax1.set_title(f'Input Digit\nTrue Label: {sample_label}', fontsize=12)
ax1.axis('off')

# Plot the probability bar chart
digits = range(10)
colors = ['red' if i == prediction else 'skyblue' for i in digits]
bars = ax2.bar(digits, probabilities, color=colors)
ax2.set_xlabel('Digit', fontsize=12)
ax2.set_ylabel('Probability', fontsize=12)
ax2.set_title(f'Prediction Probabilities\nPredicted: {prediction} (Confidence: {confidence:.2f})', fontsize=12)
ax2.set_xticks(digits)
ax2.set_ylim([0, 1])

# Add value labels on top of bars
for bar, prob in zip(bars, probabilities):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{prob:.2f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Print numerical probabilities
print("\nProbabilities for each digit:")
for digit, prob in enumerate(probabilities):
    print(f"Digit {digit}: {prob:.4f}")
```

```python
#@title 16. Upload and Predict Your Handwritten Digit Image
# ==================================================
from google.colab import files
from PIL import Image
import cv2
from scipy import ndimage
from skimage.morphology import dilation, square

def thicken_stroke(image, kernel_size=3, iterations=1):
    """
    Thicken the strokes in the image using morphological dilation
    """
    # Create kernel for dilation
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    # Apply dilation to thicken strokes
    thickened = cv2.dilate(image, kernel, iterations=iterations)

    return thickened

def preprocess_image_with_thickening(image_path, target_size=(28, 28), thicken=True, kernel_size=3, iterations=1):
    """
    Preprocess uploaded image with optional stroke thickening
    """
    # Read the image
    if image_path.endswith('.jpg') or image_path.endswith('.jpeg') or image_path.endswith('.png'):
        # Read with OpenCV
        img = cv2.imread(image_path)
        # Convert to grayscale
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        # Use PIL for other formats
        img = Image.open(image_path).convert('L')
        img = np.array(img)

    # Store original for display
    original_display = img.copy()

    # Invert colors if needed (MNIST has white digits on black background)
    if np.mean(img) > 127:
        img = 255 - img
        original_display = 255 - original_display

    # Apply threshold to make it binary
    _, img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Thicken the stroke if requested
    if thicken:
        img = thicken_stroke(img, kernel_size, iterations)

    # Store pre-thickened version for display
    before_resize = img.copy()

    # Resize to 28x28
    img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)

    # Apply threshold again after resize to maintain binary nature
    _, img = cv2.threshold(img, 63, 255, cv2.THRESH_BINARY)

    # Normalize to [0, 1]
    img = img.astype(np.float32) / 255.0

    # Add channel dimension and convert to tensor
    img_tensor = torch.from_numpy(img).float()
    img_tensor = img_tensor.unsqueeze(0)  # Add channel dimension

    # Apply MNIST normalization
    transform = transforms.Normalize((0.1307,), (0.3081,))
    img_tensor = transform(img_tensor)

    return img_tensor, img, original_display, before_resize

def predict_uploaded_image_with_thickening(model, device):
    """
    Upload an image and predict the digit with stroke thickening options
    """
    print("Please upload an image of a handwritten digit...")
    uploaded = files.upload()

    # Thickening parameters - you can adjust these
    use_thickening = True
    kernel_size = 5  # Size of thickening kernel (odd number)
    iterations = 3   # Number of thickening iterations

    for filename in uploaded.keys():
        print(f"\nProcessing file: {filename}")
        print(f"Stroke thickening: {'ON' if use_thickening else 'OFF'} (kernel={kernel_size}, iterations={iterations})")

        # Preprocess the image with thickening
        img_tensor, processed_img, original_display, before_resize = preprocess_image_with_thickening(
            filename,
            thicken=use_thickening,
            kernel_size=kernel_size,
            iterations=iterations
        )

        # Make prediction
        model.eval()
        with torch.no_grad():
            img_tensor_batch = img_tensor.unsqueeze(0).to(device)  # Add batch dimension
            output = model(img_tensor_batch)
            probabilities = torch.softmax(output, dim=1)
            predicted_class = torch.argmax(probabilities, dim=1).item()
            confidence = probabilities[0][predicted_class].item()
            all_probs = probabilities[0].cpu().numpy()

        # Display results with 4 panels
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))
        ax1, ax2, ax3, ax4, ax5, ax6 = axes.flatten()

        # Show original uploaded image
        ax1.imshow(original_display, cmap='gray')
        ax1.set_title('1. Original Uploaded Image', fontsize=12, fontweight='bold')
        ax1.axis('off')

        # Show image after thresholding (before thickening)
        ax2.imshow(before_resize, cmap='gray')
        ax2.set_title('2. After Thresholding\n(Before Thickening)', fontsize=12, fontweight='bold')
        ax2.axis('off')

        # Show image after thickening (before resize)
        if use_thickening:
            # Create a version with thickening for display
            thickened_display = thicken_stroke(before_resize, kernel_size, iterations)
            ax3.imshow(thickened_display, cmap='gray')
            ax3.set_title(f'3. After Stroke Thickening\n(kernel={kernel_size}, iter={iterations})',
                         fontsize=12, fontweight='bold')
        else:
            ax3.imshow(before_resize, cmap='gray')
            ax3.set_title('3. Thickening: OFF', fontsize=12, fontweight='bold')
        ax3.axis('off')

        # Show final 28x28 MNIST-style image
        mnist_img = processed_img
        ax4.imshow(mnist_img, cmap='gray')
        ax4.set_title('4. Final 28x28 Input\nto Model', fontsize=12, fontweight='bold')
        ax4.axis('off')

        # Plot probability bar chart
        digits = range(10)
        colors = ['red' if i == predicted_class else 'skyblue' for i in digits]
        bars = ax5.bar(digits, all_probs, color=colors)
        ax5.set_xlabel('Digit', fontsize=12)
        ax5.set_ylabel('Probability', fontsize=12)
        ax5.set_title(f'5. Prediction Probabilities\nPredicted: {predicted_class}', fontsize=12, fontweight='bold')
        ax5.set_xticks(digits)
        ax5.set_ylim([0, 1])

        # Add value labels on bars
        for bar, prob in zip(bars, all_probs):
            height = bar.get_height()
            if height > 0.01:
                ax5.text(bar.get_x() + bar.get_width()/2., height,
                        f'{prob:.2%}', ha='center', va='bottom', fontsize=8)

        # Add confidence score in a separate text box
        ax6.text(0.5, 0.6, f'PREDICTED DIGIT:\n{predicted_class}',
                fontsize=24, fontweight='bold', ha='center', transform=ax6.transAxes)
        ax6.text(0.5, 0.3, f'Confidence:\n{confidence:.2%}',
                fontsize=18, ha='center', transform=ax6.transAxes,
                bbox=dict(boxstyle='round', facecolor='lightgreen' if confidence > 0.8 else 'lightyellow', alpha=0.8))
        ax6.set_xlim(0, 1)
        ax6.set_ylim(0, 1)
        ax6.axis('off')
        ax6.set_title('6. Final Result', fontsize=12, fontweight='bold')

        plt.suptitle(f'Handwritten Digit Recognition Pipeline', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()

        # Print detailed results
        print(f"\n{'='*60}")
        print(f"PREDICTION RESULTS")
        print(f"{'='*60}")
        print(f"Predicted Digit: {predicted_class}")
        print(f"Confidence: {confidence:.2%}")
        print(f"\nTop 3 Predictions:")

        # Get top 3 predictions
        top3_idx = np.argsort(all_probs)[-3:][::-1]
        for i, idx in enumerate(top3_idx):
            print(f"  {i+1}. Digit {idx}: {all_probs[idx]:.2%}")

        # Add thickening suggestion based on results
        if confidence < 0.7 and not use_thickening:
            print(f"\n💡 Tip: Low confidence detected. Try enabling stroke thickening for better results!")
        elif confidence < 0.7 and use_thickening:
            print(f"\n💡 Tip: Try adjusting thickening parameters (increase kernel_size or iterations)")

        return predicted_class, confidence, all_probs

# Load Model and Run Prediction
best_model = DigitRecognizerNN().to(device)
best_model.load_state_dict(torch.load('best_mnist_model.pth', map_location=device))
print("Best model loaded successfully!")
print("\n" + "="*60)
print("Standard Prediction Using Your Handwritten Digit")
print("="*60)
predicted_digit, confidence, probabilities = predict_uploaded_image_with_thickening(best_model, device)
```
