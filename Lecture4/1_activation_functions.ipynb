```python
#@title Import dependencies

import numpy as np
import matplotlib.pylab as plt
```

```python
#@title Define a Step function

def step(x):
    return np.array(x > 0)
```

```python
#@title Plot the Step function

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # set y range
plt.xlabel("x")
plt.ylabel("y")
plt.title("Step function")
plt.show()
```

```python
#@title Define a Sigmoid function

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
#@title Plot the Sigmoid function

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # set y range
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid function")
plt.show()
```

```python
#@title Define a ReLU function

def relu(x):
    return np.maximum(0, x)
```

```python
#@title Plot the ReLU function

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 5.1) # set y range
plt.xlabel("x")
plt.ylabel("y")
plt.title("ReLU function")
plt.show()
```
